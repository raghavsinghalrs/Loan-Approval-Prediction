```python
df.replace(to_replace='3+', value=4,inplace=True)
df.replace(to_replace='0', value=0,inplace=True)
df.replace(to_replace='1', value=1,inplace=True)
df.replace(to_replace='2', value=2,inplace=True)
```

```python
def mis_cat(df):
    cols = ['Gender','Married','Self_Employed','Credit_History']
    for i in cols:
        df[i].fillna(df[i].mode()[0],inplace=True)
    return df
mis_cat(df)
```

```python
df.isnull().sum()
```

```python
def mis_cont(df):
    cols = ['Dependents','LoanAmount','Loan_Amount_Term']
    for i in cols:
        df[i].fillna(df[i].mode()[0],inplace=True)
    return df
mis_cont(df)
        
```

```python
df.isnull().sum()
```

```python
df.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1},
                      'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0},'Loan_Status':{'Y':1 , 'N':0}},inplace=True)
```

```python
del df['Loan_ID']
del df['Loan_Amount_Term']
```

```python
df.head(10)
```

```python
df['Loan_Status'].value_counts()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn import svm
```

```python
classifier = svm.SVC(kernel='linear',C=0.1)
classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
pred = classifier.predict(x_test)
accuracy = accuracy_score(pred,y_test)
print("Accuracy of this model is" , accuracy)
```

```python
import pickle
```

```python
filename = "trained_model.sav"
pickle.dump(classifier,open(filename,'wb'))
```

```python
loaded_model = pickle.load(open('trained_model.sav','rb'))
```

```python
input_data = (0,1,1.0,0,1,4500,1100.0,101.0,0.0,1)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = loaded_model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 1):
  print('Loan is approved')
else:
  print('Loan is not approved') 
```
